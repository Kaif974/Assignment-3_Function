Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)
n_samples = 100
feature_1 = np.random.rand(n_samples)
feature_2 = feature_1 * 0.5 + np.random.rand(n_samples) * 0.5
feature_3 = feature_1 * 0.3 + feature_2 * 0.3 + np.random.rand(n_samples) * 0.4
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})
```

```python
# Compute the correlation matrix
correlation_matrix = data.corr()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

# Create the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population'], alpha=0.5, c='blue', edgecolors='w')

# Add annotations for each country
for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]), fontsize=9, ha='right')

# Set the labels and title
plt.title('Bubble Chart of Country Population vs GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.grid(True)

# Show the plot
plt.show()
```

Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine of x

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, line_width=2, color='blue')

# Show the plot
output_notebook()  # Use this in a Jupyter notebook
show(p)
```

Q3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', height=400, width=600)

# Add bars
p.vbar(x='fruits', top='counts', width=0.4, source=source, fill_color='navy', line_color='white')

# Show the plot
output_notebook()  # Use this in a Jupyter notebook
show(p)
```

Q4.Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Set up Bokeh output
output_notebook()

# Create a figure for the histogram
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           tools='save',
           background_fill_color="#fafafa")

# Create a bar renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black')

# Show the plot
show(p)
```

Q5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

# Generate random data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid
xx, yy = np.meshgrid(x, y)

# Prepare data for the heatmap
heatmap_data = data_heatmap.flatten()
x_coords = xx.flatten()
y_coords = yy.flatten()

# Set up Bokeh output
output_notebook()

# Create a figure for the heatmap
p = figure(title="Heatmap of Random Data",
           x_axis_label='X Axis',
           y_axis_label='Y Axis',
           tools='save',
           width=400, height=400)

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(heatmap_data), high=max(heatmap_data))

# Create a source for the data
from bokeh.models import ColumnDataSource
source = ColumnDataSource(data=dict(x=x_coords, y=y_coords, value=heatmap_data))

# Add rectangles to the figure
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, line_color='white', fill_color=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
